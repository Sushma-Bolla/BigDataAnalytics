### Dataset Insights

The dataset contains details of 1,000 employees, capturing essential attributes that influence salary: Age, Gender, Education Level, Job Title, Years of Experience, and Salary. It is clean, consistent, and ready for analysis, making it ideal for exploring salary patterns and trends.

1. Salary Distribution                                                                                                                       
Salaries vary widely, from entry-level packages to top-level salaries, reflecting differences in roles, experience, and qualifications.
The average salary is around the mid-range, while the median salary closely aligns with the mean, suggesting a fairly balanced distribution.
A few high-salary outliers indicate executive or highly specialized positions.
Visual analysis shows that most employees earn a salary within the moderate range, while experience and education largely determine higher earnings.

2. Factors Affecting Salary
Several factors in the dataset correlate with salary:                                                                                        
Years of Experience: Salary increases steadily with experience. Junior employees (<5 years) earn less, mid-level (5â€“10 years) earn moderately, and senior employees (>10 years) earn the highest.                                                                               
Education Level: Higher qualifications generally lead to better compensation. Employees with advanced degrees often occupy senior or specialized roles with higher pay.                                                                                                           
Job Title: Specialized or managerial roles command higher salaries compared to entry-level positions.                                        
Age: Older employees tend to earn more, consistent with their experience and seniority.                                                      
Gender: Gender distribution is balanced, allowing analysis of salary differences across male and female employees if required.

3. Workforce Composition and Salary Insights                                                                                                 
Experience Levels: The workforce includes a mix of junior, mid-level, and senior employees, which helps in analyzing salary growth across career stages.                                                                                                                               
Education Diversity: Varying education levels allow insights into how academic qualifications influence salary.                              
Role Diversity: Multiple job titles provide a range of salary levels, highlighting the pay structure across different roles.                 
This composition makes the dataset ideal for predicting salary, understanding compensation trends, and performing HR analytics.

4. Key Salary Patterns
Salary vs Experience: Strong positive correlation; more experienced employees earn higher salaries.                                          
Salary vs Education: Higher education usually results in higher pay.                                                                         
Salary vs Job Title: Specialized and senior roles have higher compensation.                                                                  
Salary Outliers: A few employees earn significantly higher salaries, likely representing executives or highly specialized positions.

5. Overall Insights
The dataset provides a comprehensive view of salary patterns within the company. It highlights how experience, education, and role specialization drive compensation. With no missing values and a balanced mix of employees across roles and experience levels, it is well-suited for salary analysis, trend identification, and predictive modeling.

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Salary").getOrCreate()
```

```python
df = spark.read.csv("Salary_Data.csv", header=True, inferSchema=True)
```

```python
# View first 10 rows
print("=== First 10 rows of dataset ===")
df.show(10)
```

```python
# Print schema of the dataset
print("=== Schema of dataset ===")
df.printSchema()
```

```python
# Count total number of rows and columns
print("=== Total no.of rows ===")
print(df.count())
print("=== List of columns ===")
print(df.columns)
```

```python
# Select specific columns
print("=== Print age and salary columns ===")
df.select("age", "salary").show(10)
```

```python
# Employees with salary > 50000
print("=== Print employees with greater than 50000 salary ===")
df.filter(df.Salary > 50000).show(10)

# Female employees with >5 years of experience
print("\n=== Print employees with greater than 5 years of experience ===")
df.filter((df.Gender == "Female") & (df.Years_of_Experience > 5)).show(10)
```

```python
# Calculate Average, minimum and maximum salary
from pyspark.sql.functions import avg, min, max

print("=== Print average, minimum, maximum salary ===")
df.agg(
    avg("salary").alias("avg_salary"),
    min("salary").alias("min_salary"),
    max("salary").alias("max_salary")
).show()
```

```python
#Calculate average salary by gender
print("=== Print average salary by gender===")
df.groupBy("gender").agg(avg("salary").alias("avg_salary")).show()
```

```python
# Group by education level
df.groupBy("Education_level").agg(avg("salary").alias("avg_salary")).show()
```

```python
# Group by Job Title
df.groupBy("Job_title").agg(
    avg("salary").alias("avg_salary"),
    max("salary").alias("max_salary"),
    min("salary").alias("min_salary")
).orderBy("avg_salary", ascending=False).show()
```

```python
# Highest to lowest
print("=== Print salary from highest to lowest ===")
df.orderBy(df.Salary.desc()).show(10)

# Lowest to highest
print("\n=== Print salary from lowest to highest ===")
df.orderBy(df.Salary.asc()).show(10)
```

```python
# print correlation
print("=== Print correlation ===")
df.corr("Years_of_Experience", "Salary")
df.corr("Age", "Salary")
```

```python
# Print new column
print("=== Create new column Salary_k ===")
df = df.withColumn("Salary_k", df.Salary/1000)
df.select("Salary", "Salary_k").show(5)
```

```python
# Print new column
print("=== Create new column Experience Level ===")
from pyspark.sql.functions import when

df = df.withColumn("Experience_Level",
                   when(df.Years_of_Experience < 5, "Junior")
                   .when((df.Years_of_Experience >= 5) & (df.Years_of_Experience < 10), "Mid-level")
                   .otherwise("Senior"))
df.select("Years_of_Experience", "Experience_Level").show(10)
```

```python
# print first 10 rows after all operations
print("=== First 10 rows ===")
print(df.show(10))
```

```python
# Histogram showing salary distribution 
pdf = df.toPandas()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Salary distribution
plt.hist(pdf['Salary'], bins=20)
plt.xlabel("Salary")
plt.ylabel("Count")
plt.title("Salary Distribution")
plt.show()
```

```python
# Bar graph showing relation between average salary and experience level
pdf.groupby("Education_Level")['Salary'].mean().plot(kind='bar')
plt.ylabel("Average Salary")
plt.show()
```

```python
# Bar graph showing number of employees by experience level
exp_count = df.groupBy("Experience_Level").count().toPandas()

plt.figure(figsize=(7,5))
sns.barplot(data=exp_count, x="Experience_Level", y="count",hue="Experience_Level", palette="cool")
plt.title("Number of Employees by Experience Level")
plt.ylabel("Count")
plt.xlabel("Experience Level")
plt.show()
```

```python
# Heat map showing correlation between numerical columns
plt.figure(figsize=(6,4))
sns.heatmap(pdf[["Age", "Years_of_Experience", "Salary"]].corr(), annot=True, cmap="Blues")
plt.title("Correlation Between Numeric Columns")
plt.show()
```

```python
# Pie chart to Group employees by Education Level
edu_dist = df.groupBy("Education_Level").count().toPandas()

plt.figure(figsize=(6,6))
plt.pie(
    edu_dist['count'],
    labels=edu_dist['Education_Level'],
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title("Distribution of Employees by Education Level")
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()
```

```python
# Scatterplot showing relation between years of experience and age
exp_age = df.select("Years_of_Experience", "Age").toPandas()

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=exp_age,
    x="Years_of_Experience",
    y="Age",
    color="darkorange",
    s=80,
    alpha=0.7
)
plt.title("Years of Experience vs Age")
plt.xlabel("Years of Experience")
plt.ylabel("Age")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
